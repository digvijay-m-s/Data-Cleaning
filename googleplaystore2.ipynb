# Data Imputation and Manipulation

### *Fill the null values* with appropriate values using aggregate function such as *mean, Mode or Median*

```python
# define a function impute_median
## this function takes arguement as series i.e. column 
## fillna is used to fill all the missing values only

def impute_median(series):
    return series.fillna(series.median())
```

```python
###Perform a transform on Rating column
### taking each value from the rating column
## then checking if there is a null value with fillna
## if it is a null value then taking the median of the column and adding it there

google_data.Rating = google_data['Rating'].transform(impute_median)
```

```python
## fill the missing categorical values with mode
### fillna will find the missing values of the 'Type' column
### type casting it with str
###.mode() will add the mode to the missing value
### value[0] , because in case there is a bi modal value then 
## out of the 2 values choosing the first value as values[0]
## incase a second value is to be taken then  - values[1]


google_data['Type'].fillna(str(google_data['Type'].mode().values[0]), inplace=True)
```

# Data visualization

```python
grp =google_data.groupby('Category') 
x= grp['Rating'].agg(np.mean)
y= grp['Price'].agg(np.sum)
z=grp['Reviews'].agg(np.mean)
m=grp['Installs'].agg(np.mean)

print(x)
print(y)
print(z)
print(m)
```

```python
plt.figure(figsize=(12,5))
plt.plot(x , 'go')
plt.xticks(rotation=90)
plt.title('Category wise rating')
plt.xlabel('Categories')
plt.ylabel('Ratings')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(y , 'r--')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Categories')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(z , 'b-')
plt.xticks(rotation=90)
plt.title('Category wise Reviews')
plt.xlabel('Categories')
plt.ylabel('Reviews')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(m , 'ro')
plt.xticks(rotation=90)
plt.title('Category wise Installs')
plt.xlabel('Categories')
plt.ylabel('Installs')
plt.show()
```
